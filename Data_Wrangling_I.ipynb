# Merge y Concat

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
 'first': range(7)})
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'd'], 'second': range(3)})
```

```python
df1
```

```python
df2
```

```python
df2.info()
```

```python
pd.merge(df1,df2, on ='key',how='left')
```

```python
pd.merge(df1,df2, on ='key',how= 'inner')
```

```python
pd.merge(df1,df2, on ='key',how= 'outer')
```

```python
pd.merge(df1,df2, on ='key',how= 'left')
```

```python
result = pd.merge(df1,df2, on ='key',how= 'right')
result
```

```python
df1.describe()
```

```python
result["first"].astype("int")
# NaN not int
```

```python
result.key.value_counts()
```

# Consultas "tipo SQL"

```python
customers= pd.read_csv('customers.csv',sep=',')
customers.head()
```

```python
employees= pd.read_csv('employees.csv',sep=',')
employees.head()
```

```python
orders= pd.read_csv('orders.csv',sep=',')
orders.head()
```

```python
rta1=pd.merge(orders, employees, on='EmployeeID', how='inner')
rta1[['OrderID', 'FirstName', 'LastName']].sort_values(by='OrderID')
```

```python
# Lo mismo que el bloque anterior, pero todo junto
pd.merge(orders, employees, how='inner', on='EmployeeID').sort_values(by='OrderID')[['OrderID','LastName','FirstName']]
```

```python
rta2 = pd.merge(orders, customers, on = 'CustomerID', how = 'inner')
rta2[['OrderID', 'CustomerName']].sort_values(by='OrderID')
```
